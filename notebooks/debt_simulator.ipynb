## *Patrimônio* após *t* meses com compra no mês *Tc*:

### Parâmetros:

Mês da compra (a partir de t=0):         Tc

Ganho mensal pela Resolução do problema: Gr

Valor de manutenção do Bem:              Vm

Taxa de preciação do Bem (mensal):       pm

#### Exemplo:

```python
Tc_exemplo           = 6                                    # Mês da compra (a partir de t=0) ; int

Gr_exemplo           = 200.00                               # Ganho mensal pela Resolução do problema ; float

Vm_exemplo           = 1500.00                              # Valor de manutenção do Bem ; float

pm_exemplo           = eq.anual_para_mensal(0.90)              # depreciação de 10% ao ano do valor do bem adquirido, ou seja, apreciação negativa de 10% ao ano ; float

Am_preparado_exemplo = Am_exemplo + Vp_exemplo + Vm_exemplo # O parâmetro de *aporte inicial* já contabiliza o valor separado no orçamento para as *parcelas e manutenção* do Bem ; float
```

#### Observações:

Uma compra é feita para resolver um problema. Caso esse **problema tenha um custo real mensal**, esse custo será um **bônus no orçamento** após a compra. Exemplo: uber -> carro; aluguel -> casa própria. Esse bônus sera adicionado ao *aporte inicial* como Gr.

Para comprar algo, é preciso ter o dinheiro necessário. Se uma pessoa está **preparada para comprar** algo, é necessário que ela tenha o orçamento para pagar as parcelas e o custo de manutenção, e é realista assumir que a pessoa terá **separado esse dinheiro no orcamento antes de comprar** o Bem (ganhou um aumento, cortou custos, se mudou pra um lugar mais barato, etc...). Normalmente a pessoa organiza o orçamento em um mês e compra no seguinte, mas, se esperasse uns meses (pesquisando ou esperando ofertas, por exemplo), ela poderia **investir esse dinheiro até realizar, efetivamente, a compra** (no mínimo, guardando o dinheiro separado nas caixinhas do Nubank ou porquinho do Inter, onde renderia com uma taxa do CDB). Por isso, o parâmetro de *aporte inicial* já contabiliza o valor separado no orçamento para as *parcelas e manutenção* do Bem.

O ideal é assumir que uma **pessoa pode tanto comprar o Bem à vista quanto financiado**, isto é, ela **separou o dinheiro no orçamento** para as parcelas do financiamento **e** tem o **dinheiro necessário em conta** para o pagamento à vista. Veremos que a escolha correta nem sempre é o pagamento à vista, a depender dos parâmetros iniciais da pessoa. Obviamente, caso a pessoa não tenha o valor necessário para pagar à vista, a única escolha é empréstimo, mas **esse simulador ainda serve para comparar quais condições de empréstimo são as mais favoráveis, mesmo que o agiota acorde de mal humor**.

O **Bem** tem um valor que pode ser **adicionado ao patrimônio ao final** do *Horizonte*. Porém, o valor de **Bens diferentes oscilam de maneiras diferentes**. Por exemplo, uma **casa pode aumentar** de valor com o tempo, aumentando o patrimônio efetivo do proprietário, já um **carro quase sempre diminui** de valor com o tempo. Isso entra no modelo como uma *taxa de preciação*, que pode ser **positiva (apreciação)** ou **negativa (depreciação)**, que vai diminuir ou aumentar o Bem de valor ao longo do tempo.

O *aporte mensal* continua *aumentando* durante o endividamento.

### Relações Temporais:

**0 < Tc < (Tc + np) < Th** , para todos Tc,np e Th pertencentes aos **números naturais positivos**.

#### Conclusão:

A simulação, *teóricamente*, independe da relação temporal entre os parâmetros. Isto é, se *Tc > Th*, a simulação deve correr como se a pessoa tivesse mudado de ideia e não comprado o Bem, por exemplo. Simulação necessária para comprovação ou disprovação dessa teoria.

### Defeito de Domínio:

As equações dividem o tempo em **3 períodos**: Antes, Durante e Após a compra. É possível chegar à conclusão, analisando as equações, que há **intersecção entre os períodos**, isto é, o mês de divisa é contabilizado duas vezes. Como pensamos, idealmente, em um *Horizonte* muito grande (10 anos, 120 meses, por exemplo), **dois meses extras não adicionariam erro grande no resultado final**, não tanto quanto as eleições presidenciais adicionariam erro nas previsões econômicas, pelo menos. Também, a efeito de comparação, teóricamente, **não há adição de vantagem** na simulação para **empréstimo ou à vista**, pois a **matemática é cruel e não perdoa ninguém**.

Porém, é de se notar a **diferença entre os casos**, tendo em vista esse erro equacional:

#### Empréstimo:

Mês Tc: As equações investem 1 *Aporte Resulante* a mais, que renderá até o *Horizonte*.

Mês (Tc + np): As equações investem mais 1 *Aporte Resulante* a mais, que renderá até o *Horizonte*.

Resultado: 1 *aporte* do Mês *Tc* + 1 *aporte* do Mês *(Tc + np)* adicionais.

#### À Vista:

Mês Tc: As equações investem 2 *aportes* desse mês a mais, que renderão até o *horizonte*.

Resultado: 2 *aportes* do mês *Tc* adicionais.

#### Conclusão:

Na simulação de **Empréstimo**, o investimento extra de *Tc* rende por *(Th - Tc)* meses e o de *(Tc + np)* rende por *[Th - (Tc + np)]* meses, que é menos tempo. Já na de **À Vista**, os dois investimentos em *Tc* rendem por *(Th - Tc)*. Isso pode ser uma **vantagem equacional** para o modo à vista. Porém, é de se notar que o *aporte* extra do mês *(Tc + np)* é **maior que o aporte do mês *Tc*** e, no fim, a escala de tempo deve tornar essas **diferenças muito pequenas**.

De todo modo, simulações são necessárias para conclusão precisa. 

### Grandes Horizontes:

Por vezes, ao simular *tempos de horizonte* **muito grandes**, este simulador calcula valores, *aparentemente*, absurdos. 

*Como assim terei **20 milhões** investindo só 500tinho por mês? Tem algum erro nisso, certeza!*

A resposta para tal comentário é que **não há erro**, somente a matemática dos **juros compostos** mostrando seu poder.

*Se fosse fácil assim **qualquer um** fazia!*

A resposta à essa resposta àquela resposta ao comentário é que **sim**, *se fosse fácil qualquer um fazia*. Resultados assim exigem dois ingredientes *preciosíssimos* no mundo: **Tempo e Consistência**.

Pode *começar a investir desde os dois anos de idade*, mas se *perder tanto dinheiro quanto render*, morrerá de velhice no **zero**.
Analogamente, pode *conseguir rendimentos que dobrem seu dinheiro* a cada dois anos (**o sonho de qualquer investidor!**), mas se morrer de velhice em *10 anos*, só será um milionário caso já tenha *começado com grana alta* no bolso.
